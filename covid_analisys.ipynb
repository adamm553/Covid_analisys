# ZaleÅ¼noÅ›Ä‡ pomiÄ™dzy iloÅ›ciÄ… zgonÃ³w, a GDP per capita

```python
gdp = covid.groupby('location').max()['gdp_per_capita']
deaths = covid.groupby('location').max()['total_deaths']
```

```python
df1 = pd.DataFrame(gdp)
df2 = pd.DataFrame(deaths)
df = pd.concat([df1, df2])
round(df.describe()['total_deaths'], 1)
```

# Åšrednia liczba wszystkich zgonÃ³w wynosi 30545.3
# Liczba zgonÃ³w odchyla siÄ™ od Å›redniej o 103296.4
# Maksymalna liczba zgonÃ³w wynosi 1056416.9

```python
plt.scatter(gdp.values, deaths.values, c='green')
plt.ylim((10, 100000))
plt.xlabel("GDP per capita")
plt.ylabel("Zgony")
plt.show()
```

# Wraz ze wzrostem zamoÅ¼noÅ›ci spoÅ‚eczeÅ„stwa, liczba zgonÃ³w maleje

# WpÅ‚yw szczepieÅ„ na iloÅ›Ä‡ zakaÅ¼eÅ„

```python
covid['date_month'] = covid['date'].str[0:7]
```

```python
vl = covid.groupby('date_month').sum()['new_cases']
vl1 = covid.groupby('date_month').max()['new_vaccinations']
```

```python
plt.plot(vl.index, vl)
plt.plot(vl1.index, vl1)
plt.xlabel('MiesiÄ…c i rok')
plt.xticks(rotation='vertical')
plt.ylabel("Liczba nowych zakaÅ¼eÅ„ oraz szczepieÅ„ (w milionach)")
plt.show()
```

# Szczepienia wpÅ‚ynÄ™Å‚y na ustabilizowanie nowych zakaÅ¼eÅ„ po okoÅ‚o roku czasu

# RozwÃ³j pandemii COVID-19 w Polsce

```python
wock_to_poland = covid[covid['location']=='Poland']
wock_to_poland['month_year'] = wock_to_poland['date'].str[0:7]
wock_to_poland.head()
```

```python
mont = wock_to_poland.groupby('month_year').sum()
values = mont['new_cases']
```

```python
plt.plot(values.index, values)
plt.xlabel('MiesiÄ…c i rok')
plt.xticks(rotation='vertical')
plt.ylabel("Liczba nowych zakaÅ¼eÅ„")
plt.show()
```

# IloÅ›Ä‡ robionych testÃ³w w Polsce w porÃ³wnaniu do innych krajÃ³w

```python
poland = covid[covid['location']=='Poland']
poland = poland[['new_vaccinations', 'date_month']]
maks1 = poland['new_vaccinations'].sum()
```

```python
niemiec = covid[covid['location']=='Germany']
niemiec = niemiec[['new_vaccinations', 'date_month']]
maks2 = niemiec['new_vaccinations'].sum()
```

```python
fran = covid[covid['location']=='France']
fran = fran[['new_vaccinations', 'date_month']]
maks3 = fran['new_vaccinations'].sum()
```

```python
x = poland.groupby('date_month').max()
plt.xticks(rotation='vertical')
plt.plot(x.index, x, label='Polska')
y = niemiec.groupby('date_month').max()
plt.xticks(rotation='vertical')
plt.plot(y.index, y, label='Niemcy')
z = fran.groupby('date_month').max()
plt.plot(z.index, z, label='Anglia')
plt.legend()
plt.show()
```
