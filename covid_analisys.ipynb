# Minimalny i maksymalny stosunek iloÅ›ci zgonÃ³w do iloÅ›ci zakaÅ¼eÅ„ 

```python
covid['pct'] = (covid['total_deaths'] / covid['total_cases'])*100 
```

```python
covid
```

```python
con_min = round(covid.groupby('location').min(), 2)['pct']
con_max = round(covid.groupby('location').max(), 2)['pct']
datas_min = con_min[['Africa', 'Asia','Europe', 'South America', 'North America', 'Oceania']]
datas_max = con_max[['Africa', 'Asia','Europe', 'South America', 'North America', 'Oceania']]
```

```python
labels = datas_min.index
x = np.arange(len(labels)) 
width = 0.35
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, datas_min, width, label='minimalny stosunek')
rects2 = ax.bar(x + width/2, datas_max, width, label='maksymalny stosunek')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Minimalny i maksymalny stosunek iloÅ›ci zgonÃ³w do iloÅ›ci zakaÅ¼eÅ„ na kontynencie')
ax.set_xticks(x, labels, rotation='vertical')
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

# NajwiÄ™ksze zrÃ³Å¼nicowanie zgonÃ³w do zakaÅ¼eÅ„ byÅ‚o zaobserwowane w Europie. Tam teÅ¼ roÅ¼nica pomiÄ™dzy najwiÄ™kszym, a najmniejszym stosunkiem jest najwiÄ™ksza

# RozwÃ³j pandemii COVID-19 w Polsce

```python
wock_to_poland = covid[covid['location']=='Poland']
wock_to_poland['month_year'] = wock_to_poland['date'].str[0:7]
wock_to_poland.head()
```

```python
mont = wock_to_poland.groupby('month_year').sum()
values = mont['new_cases']
```

```python
plt.plot(values.index, values)
plt.xlabel('MiesiÄ…c i rok')
plt.xticks(rotation='vertical')
plt.ylabel("Liczba nowych zakaÅ¼eÅ„")
plt.show()
```
